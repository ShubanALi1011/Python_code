**8. Write a program that reads in the radius and height of a cylinder and computes the
area and volume.**

```python
import math

radius = float(input("ENTER RADIUS OF CYLINDER : "))
height = float(input("ENTER HEIGHT OF CYLINDER : "))

area = (2 * math.pi * (radius ** 2)) + (2 * math.pi * radius * height)
volume = math.pi * (radius ** 2) * height

print(f"THE AREA OF CYLINDER IS {area}")
print(f"THE VOLUME OF CYLINDER IS {volume}")
```

**3. Write a program that prompts the user to enter three integers and displays the numbers
in reverse order. The expected output is shown below:**

```python
a = int(input("ENTER FIRST NUBER : "))
b = int(input("ENTER FIRST NUBER : "))
c = int(input("ENTER FIRST NUBER : "))

maxi = max(a,b,c)
mini = min(a,b,c)
mid = a + b + c - maxi - mini

print(f"NUMBERS in Sorted Order {mini,mid,maxi}")
```

**4. Write a program that prompts the user to enter the side of a hexagon and displays
its area. The formula for computing the area of a hexagon is written below, where s
is the length of a side.**

```python
import math
s = float(input("ENTER SIDE OF HEXAGON : "))
area = ((3 * math.sqrt(3)) / 2 )* (s ** 2)
print(f"The AREA OF HEXAGON IS {area}")
```

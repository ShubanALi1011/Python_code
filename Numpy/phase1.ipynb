## Numpy Array & Basics


```python
import numpy as np 
```

## Creating Array From List

```python
array_id = np.array([1,2,3,4,5])
print("1D Array : ",array_id)

arr_2d = np.array([[1,2,3],[4,5,6]])
print("2D Array : ",arr_2d)
```

## List vs numpy array

```python
py_list = [1,2,3,4]
print("Python List Multiplication : ",py_list*2)
np_array = np.array([1,2,3,4])
print("Numpy Array Multiplication : ", np_array*2)
```

Checking Time

```python
py_list = [1,2,3,4]
print("Python List Multiplication : ",py_list*2)
np_array = np.array([1,2,3,4])
print("Numpy Array Multiplication : ", np_array*2)

import time
start = time.time()
py_list = [i*2 for i in range(1000000)]
print("\n List Operation Time : ",time.time()-start)

start = time.time()
np_array = np.arange(1000000) * 2
print("\n Numpy Operation Time : ",time.time()-start)
```

Creating Array From Scratch

```python
zeros = np.zeros((3,4))
print("Zero Arrays :-\n",zeros)
print("\n")
ones = np.ones((3,4))
print("Ones Arrays :-\n",ones)
print("\n")

full = np.full((3,4), 3)
print("Full Arrays :-\n",full)

print("\n")

random = np.random.random((2,3))
print("Random Array :-\n ",random)

sequence = np.arange(0,11, 2)
print("Sequence : ",sequence)
```

Vector, Matrix and Tensor

```python
vector = np.array((1,2,3))
print("vector : ",vector)

print("\n")

matrix = np.array([[1,2,3], [4,5,6]])
print("Matrix :-\n",matrix)

print("\n")

tensor = np.array([
    [[1, 2], [3, 4]],
    [[5, 6], [7, 8]]
])
print("Tensor :-\n",tensor)
```

Properties of Array

```python
arr = np.array([[1,2,3],
                [4,5,6]])
print("Shape : ", arr.shape)
print("Dimension : ", arr.ndim)
print("Size : ", arr.size)
print("Data Type : ", arr.dtype)
```

Array Reshaping

```python
arr = np.arange(12)
print("Original Array : ", arr)

reshaped = arr.reshape((3,4))
print("\nReshaped Array :-\n ", reshaped)

flattern = reshaped.flatten()
print("\nFlattern :-\n", flattern)

# ravel (return view, instead of copy)
reveled = reshaped.ravel()
print("\nReveled Array : ", reveled)

#transpose
transpose = reshaped.T
print("\nTranspose of Reshaped :-\n ", transpose)
```

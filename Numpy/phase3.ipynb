## Advance Operation with Business Examples


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data Structure : [restuarent_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 15000, 180000, 220000, 250000], # Paradise Biryani
    [2, 120000, 140000, 160000, 190000], # Beijing Bites
    [3, 200000, 230000, 260000, 300000], # Pizza Hub
    [4, 180000, 210000, 240000, 270000], # Burger Point
    [5, 160000, 185000, 205000, 230000], # Chai Point
])

print(" === Alien Sales Analysis ===")
print("\nSales Data Shape : ",sales_data.shape)
print("\n Sample Data for first 3 res : \n", sales_data[:3])
```

```python
# Total Sales Per Year
print(np.sum(sales_data, axis = 0))
yearly_total = np.sum(sales_data[:, 1:], axis = 0) # axis 0 col wise
print(yearly_total)
```

```python
# Minimum Sales Per Res
min_sales = np.min(sales_data[:,1:], axis=1)
print(min_sales)
```

```python
max_sales = np.max(sales_data[:, 1:], axis = 0)
print(max_sales)
```

```python
# Avg Sales Per Res
avg_sales = np.average(sales_data[:, 1:], axis = 1)
print(avg_sales)
```

```python
# Cummulative Sales
cumsum = np.cumsum(sales_data[:, 1:], axis = 1)
print(cumsum)

plt.Figure(figsize= (10, 6))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Average Cummulative Sales")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.grid(True)
plt.show
```

```python
vector1 = np.arange(1, 6)
vector2 = np.arange(6, 11)

print("Vector 1 is : ", vector1)
print("Vector 2 is : ", vector2)

print("\nVector sum is : ", vector1 + vector2)
print("\nVector Multiplication : ", vector1 * vector2)
print("\nDot Product : ", np.dot(vector1, vector2))

# Correct angle formula
angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print("\nAngle between vectors (in radians): ", angle)

# Vectorize string operation
res_types = np.array(['biryani', 'chinese', 'pizza', 'burger', 'cafe'])
vector_upp = np.vectorize(str.upper)
print("\nVectorize Upper : ", vector_upp(res_types))
```

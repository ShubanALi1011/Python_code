# Select Specific Column and Filter rows


# Problems
* select specific column( using sq brackets )
* filter rows  ( using boolen )
* combine multiple conditions






```python
columns = df['Names']
print(columns)
```

```python
name_salary = df[['Names', 'Salary']]
print(name_salary)
```

```python
filtered = df[(df['Salary'] > 70000) & (df['Salary'] < 200000)]
print(filtered)
```

```python
print(df[df['Names'] == 'Shuban'])
```

```python
# salary > 50k and age is less than 20
filtered = df[(df['Age'] < 20) & (df['Salary'] > 50000)]
print(filtered)
```

```python
# salary < 100k and age is less than 20
filtered = df[(df['Age'] < 20) | (df['Salary'] < 65000)]
print(filtered)
```

```python
import pandas as p
```

```python
data = {
    "Names" : ['Shuban', 'Taimoor', 'Huzaifa', 'Hadia', 'Ali', 'Muneeb', 'Tuba', 'Warda'],
    "Age" : [19,20,18,21,22,20, 21, 22],
    "Salary" : [250000, 200000, 100000, 80000, 90000, 100000, 50000, 60000],
    "Performance" : [99,95,85,80,85,90,70,75]
}
df = p.DataFrame(data)
print(df)

# Adding col via assign
df["Destination"] = ['ML Engineer', 'CEO', 'COO', 'Software Engineer', 'Full Stack', 'Full Stack', 'Accountant', 'Human Resouce']
print("\n\n")
print(df)

# cal col
df["Bonus(10%)"] = df['Salary'] * 0.1
print("\n\n")
print(df.head(8))
```

```python
# add using insert
# df.insert(loc, "Column Name", "Data")
df.insert(0,"Emp_ID", ["25-Alien-001","25-Alien-002","25-Alien-003","25-Alien-004","25-Alien-005", "25-Alien-006", "25-Alien-007", "25-Alien-008"])
print(df)
```

```python
# update salary of hadia
df.loc[3,"Salary"] = 120000
print(df)
```

```python
# increase salary 20% for each
df["Salary"] += df["Salary"]  * 0.2
print(df)
```

```python
## Remove columns
df.drop(columns=['Performance'], inplace=True)
print(df)
```
